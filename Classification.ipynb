```python
# OTP messages
```

```python
unique=df[df["Category"]=="Spam"]["Content"]
unique_index=df[df["Category"]=="Spam"]["Content"].shape[0]


ll=[]
p=re.compile("\d{4,6}")
for w,x in zip(range(unique_index),unique):
    if type(x)==str:
        findings=p.findall(x)
        if "OTP" in x and len(findings)==1:
            ll.append(w)
            print(x,"\n")     
```

###  Converting to OTP

```python
# Converting to "OTP"
import time
start=time.time()
for i in range(df.shape[0]):
    for j in range(len(ll)):
        if i==j:                  # using indices
            df["Category"][i]="OTP"
#             print(df["Category"][i],"\n")
end=time.time()
print(end-start)
```
